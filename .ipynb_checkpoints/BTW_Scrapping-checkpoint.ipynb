```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
def get_soup(url):
    responce=requests.get(url)
    html=responce.text
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
website="https://www.btwonline.com/"
soup_site=get_soup(website)
```

```python
main_div=soup_site.find('div',attrs={'id':'NavigationTier2-4'})
```
